```python
import numpy as np
import cv2
from scipy import ndimage
from skimage import io
import matplotlib.pyplot as plt
```

```python
img = np.load("C:/Users/matiu/Downloads/CFA_NPY/CFA_NPY/Fuji/panda.npy")
height, width, c = img.shape
img2 = img.copy()
```

```python
io.imshow(img)
```

```python
#Mozaikowanie
for h in range(height):
    for w in range(width):
        pixel = img2[h][w]
        if h % 2 == 0 and w % 2 == 0:
            pixel[0] = 0
            pixel[2] = 0
        elif h % 2 == 0 and w % 2 == 1:
            pixel[0] = 0
            pixel[1] = 0
        elif h % 2 == 1 and w % 2 == 0:
            pixel[1] = 0
            pixel[2] = 0
        else:
            pixel[0] = 0
            pixel[2] = 0
io.imshow(img2)
```

```python
#Demozaikowanie
for h in range(height):
    for w in range(width):
        pixel = img2[h][w]
        if h % 2 == 1 and w > 0 and w < width - 1 and pixel[0] == 0:
            pixel[0] = img2[h][w - 1][0] / 2 + img2[h][w + 1][0] / 2
        if h % 2 == 0 and w > 0 and w < width - 1 and pixel[2] == 0:
            pixel[2] = img2[h][w - 1][2] / 2 + img2[h][w + 1][2] / 2
        if w > 0 and w < width - 1 and pixel[1] == 0:
            pixel[1] = img2[h][w - 1][1] / 2 + img2[h][w + 1][1] / 2
        if w % 2 == 0 and h > 0 and h < height - 1 and pixel[0] == 0:
            pixel[0] = img2[h - 1][w][0] / 2 + img2[h + 1][w][0] / 2
        if w % 2 == 1 and h > 0 and h < height - 1 and w > 0 and w < width - 1 and pixel[0] == 0:
            pixel[0] = img2[h - 1][w - 1][0] / 2 + img2[h + 1][w + 1][0] / 2
        if w % 2 == 1 and h > 0 and h < height - 1 and pixel[2] == 0:
            pixel[2] = img2[h - 1][w][2] / 2 + img2[h + 1][w][2] / 2
        if w % 2 == 0 and h > 0 and h < height - 1 and w > 0 and w < width - 1 and pixel[2] == 0:
            pixel[2] = img2[h - 1][w - 1][2] / 2 + img2[h + 1][w + 1][2] / 2
```
